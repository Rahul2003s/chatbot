```python
def read_tamil_data(file_name):
    with open(file_name, 'r', encoding='utf-8') as file:
        tamil_text = file.read()
    line_by_line = tamil_text.split("\n\n")
    tamil_data = []
    for line in line_by_line:
        question, answer = line.split(":-",1)
        tamil_data.append({"question":question,"answer":answer})
    return tamil_data
```

```python
def write_tamil_data_to_csv(tamil_data, csv_file_name):
    with open(csv_file_name, 'w', newline='', encoding='utf-8') as csv_file:
        writer = csv.DictWriter(csv_file, fieldnames=['question', 'answer'])
        writer.writeheader()
        writer.writerows(tamil_data)
```

```python
tamil_data = read_tamil_data('test.txt')
len(tamil_data)
```

```python

write_tamil_data_to_csv(tamil_data, './output.csv')
```

```python
def read_csv(file_name):
    with open(file_name, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        data = [row for row in reader]

    return data

```
